```python
import pandas as pd

# Dataset:
data = {
    'CustomerID': [1, 2, 1, 3, 2, 1, 4, 3, 5, 2, 3],
    'TransactionDate': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05',
                        '2021-01-06', '2021-01-07', '2021-01-08', '2021-01-09', '2021-01-10', '2021-01-11'],
    'ProductID': [101, 102, 101, 103, 102, 101, 104, 103, 105, 102, 103],
    'Quantity': [2, 1, 3, 2, 2, 1, 1, 1, 2, 1, 3],
    'Price': [10, 20, 10, 15, 20, 10, 25, 15, 30, 20, 15]
}

df = pd.DataFrame(data)
df
```

```python
# 1. Calculate the total amount spent by each customer.

total_spent = df.groupby("CustomerID")["Price"].sum().reset_index()
total_spent.columns = ["CustomerID", "TotalSpent"]
print(total_spent)
```

```python
# 2. * Identify the top 5 customers who have spent the most.

top_5_customers = total_spent.nlargest(5, "TotalSpent")
print(top_5_customers)
```

```python
from collections import Counter
```

```python
# 3. * For each of these top 5 customers, determine their most frequently purchased product.
def most_frequent_product(customer_id):
    customer_products = df[df["CustomerID"] == customer_id]["ProductID"]
    most_frequent = Counter(customer_products).most_common(1)[0][0]
    return most_frequent

top_5_customers = total_spent.nlargest(5, "TotalSpent")
top_5_customers["MostFrequentProduct"] = top_5_customers["CustomerID"].apply(most_frequent_product)
print(top_5_customers)
```

```python
from os import name
# 4.  Create a summary DataFrame that includes CustomerID, TotalSpent, and MostFrequentProduct for these top 5 customers."

summary_df = top_5_customers[["CustomerID", "TotalSpent", "MostFrequentProduct"]]
summary_df.columns = ["CustomerID", "TotalSpent", "MostPurchasedProduct"]
summary_df = pd.DataFrame(summary_df)

print(summary_df)
```

```python
# 1. Develop a script that takes product attributes as input.
# 2. Generate a descriptive paragraph summarizing key features and selling points.
# 3. Ensure the descriptions are grammatically correct and engaging.


import pandas as pd



# df = pd.read_csv("product_attributes.csv")


columns  = ["product_id", "name", "brand", "category", "features"]

df = pd.DataFrame(columns=columns)


from transformers import pipeline
generator  = pipeline("text-generation", model="gpt2", tokenizer="gpt2")

def generate_description(product_features):

    prompt = (f" Generate a consise and appealing product description and descriptive paragraph summarizing key features and selling points the follwing products Ensure the descriptions are grammatically correct and engaging.:\n"
                f"Name: {product_features['name']} \n"
                f"Brand: {product_features['brand']} \n"
                f"Category: {product_features['category']} \n"
                f"Features: {', '.join(product_features['features'])} \n"
                f"Description:"
    )

    generated = generator(prompt, max_length=1000, num_return_sequences=1)
    generated_text = generated[0]['generated_text']

    return generated_text

description = generate_description(df)
print(description)



# df["description"] = df.apply(generate_description, axis=1)
# df
```
